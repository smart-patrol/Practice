```python
def years_played_to_date(player):
    years_to_date = player.year - np.min(player.year) + 1
    player['years_to_date'] = years_to_date
    return player
```

```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
# columns to apply to 
cols = df.columns[0:4]
```

Measure time of apply vs Map

```python
t0 = time.clock()
df[cols].apply(lambda x : x / 2).head(2) 
print time.clock() - t0
```

```python
c_list = df.columns[0:4].tolist()
```

```python
t0 = time.clock()
for c in c_list:
    df[c].map(lambda x : x / 2).head(2)
print time.clock() -t0
```

```python
t0 = time.clock()
df['sepal length (cm)'].apply(lambda x : x / 2).head(2)
print time.clock() - t0
```

```python
t0 = time.clock()
df['sepal length (cm)'].map(lambda x : x / 2).head(2)
print time.clock() - t0
```

```python
# wide to long
df['id'] = np.arange(0, len(df))
df['species'] = pd.Categorical(iris.target, levels = iris.target_names)
df_lng = pd.melt(df, id_vars = ['id', 'species'])
print df_lng.head()
print df_lng.tail()
```

# Multi-Level Indexing

```python
from datetime import datetime
```

```python
n = 100000000
df = pd.DataFrame({
        "timestamp" : [datetime.now() for t in range(n)],
        "value" : np.random.uniform(-1,1,n)
    })
```

```python
df.head()
```

```python
df.describe()
```

```python
df.count()
```

```python
# time dataframe operation
s = time.time(); df.describe(); e = time.time()-2;
```

```python
print e
```

```python
# create an index
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['value'] = df['value'].cumsum()
df.index = df['timestamp']
```

```python
# see what index can do
```

```python
s = time.time(); df.describe() ; e = time.time() - s
```
